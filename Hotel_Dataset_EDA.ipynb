```python
import pandas as pd
```

```python
df = pd.read_csv('hotel_bookings.csv')
df
```

```python
basic_info = {
    "shape": df.shape,
    "columns": df.columns.tolist(),  # Perbaikan: Mengakses kolom dengan .columns
    "data_types": df.dtypes,
    "missing_values": df.isnull().sum(),
    "sample_data": df.head()
}

basic_info
```

```python
missing_values = df.isnull().sum()
missing_values_with_nulls = missing_values[missing_values > 0]
pd.DataFrame(missing_values_with_nulls)
```

```python
# Menghapus baris yang memiliki nilai null di kolom 'country'
hotel_bookings_cleaned = df.dropna(subset=['country'])
hotel_bookings_cleaned.head()
```

```python
hotel_bookings_cleaned.describe()
```

```python
print(hotel_bookings_cleaned.isnull().sum())
```

```python
hotel_bookings_cleaned['country'].unique()
```

```python
hotel_bookings_cleaned.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Contoh: Rata-rata 'adr' berdasarkan 'hotel' dan 'arrival_date_month'
pivot_table = hotel_bookings_cleaned.pivot_table(
    values='adr', 
    index=['arrival_date_year', 'arrival_date_month'],
    columns='hotel', 
    aggfunc='mean'
)
```

```python
# Membuat heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu")

plt.title("Average ADR by Hotel and Month")
plt.ylabel("AVG by Year Month")
plt.ylabel("Hotel Type")
plt.show()
```

```python
# Scatter plot: lead_time vs adr
plt.figure(figsize=(10, 6))
sns.scatterplot(data=hotel_bookings_cleaned, x='stays_in_week_nights', y='adr', alpha=0.6, edgecolor=None)
plt.title('stays in week nights vs Average Daily Rate (ADR)', fontsize=16)
plt.xlabel('stays in week nights', fontsize=12)
plt.ylabel('ADR (Average Daily Rate)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Menghitung Q1, Q3, dan IQR untuk stays_in_week_nights dan adr
Q1_stays = hotel_bookings_cleaned['stays_in_week_nights'].quantile(0.25)
Q3_stays = hotel_bookings_cleaned['stays_in_week_nights'].quantile(0.75)
IQR_stays = Q3_stays - Q1_stays

Q1_adr = hotel_bookings_cleaned['adr'].quantile(0.25)
Q3_adr = hotel_bookings_cleaned['adr'].quantile(0.75)
IQR_adr = Q3_adr - Q1_adr

# Menghitung batas bawah dan batas atas
lower_bound_stays = Q1_stays - 1.5 * IQR_stays
upper_bound_stays = Q3_stays + 1.5 * IQR_stays

lower_bound_adr = Q1_adr - 1.5 * IQR_adr
upper_bound_adr = Q3_adr + 1.5 * IQR_adr

# Outlier stays_in_week_nights
outliers_stays_values = hotel_bookings_cleaned[
    (hotel_bookings_cleaned['stays_in_week_nights'] < lower_bound_stays) |
    (hotel_bookings_cleaned['stays_in_week_nights'] > upper_bound_stays)
]['stays_in_week_nights']

# Outlier adr
outliers_adr_values = hotel_bookings_cleaned[
    (hotel_bookings_cleaned['adr'] < lower_bound_adr) |
    (hotel_bookings_cleaned['adr'] > upper_bound_adr)
]['adr']

```
